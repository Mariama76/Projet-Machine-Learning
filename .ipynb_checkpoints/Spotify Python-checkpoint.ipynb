## Importation des données:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Lecture des données

## Chargement les données ou les lire directement en précisant le chemin
path=""

spotify=pd.read_csv(path+"spotify-extr.txt",sep=" ",header=0)
# Vérification du contenu
spotify.head()
```

```python
#on affecte le bon type aux variables
spotify["key"]=pd.Categorical(spotify["key"],ordered=False)
spotify["pop.class"]=pd.Categorical(spotify["pop.class"],ordered=False)
spotify["mode"]=pd.Categorical(spotify["mode"],ordered=False)
spotify.dtypes
```

```python
spotify.describe()
```

```python
list1=["valence", "year", "acousticness", "danceability", "duration", "energy", "instrumentalness", "liveness", "loudness", "popularity", "speechiness", "tempo"]
fig, axs = plt.subplots(3, 4, figsize = (20,15))
plt.figure(1)
for i in range(len(list1)):
    plt.subplot(4,3,i+1)
    spotify[list1[i]].hist()
    plt.title(list1[i])
    
plt.show()
```

```python
# scatter plot matrix des variables quantitatives
from pandas.plotting import scatter_matrix
scatter_matrix(spotify[list1], alpha=0.2, figsize=(15, 15), diagonal='kde')
plt.show()
#pas interprétable
```

```python
spotify1=pd.read_csv(path+"spotify-extr.txt",sep=" ",header=0)
del spotify1["key"]
del spotify1["pop.class"]
del spotify1["mode"]
```

```python
# Correlation matrix 
spotify_data=spotify1.to_numpy()
CM = np.corrcoef(spotify_data.T)

plt.figure(figsize = (20, 20))
plt.imshow(CM, vmin=-1)

plt.title("Matrice des corrélations des variables Spotify", fontsize = 18)
plt.xticks(ticks = np.arange(0, 12, 1), labels = list1, fontsize = 15)
plt.yticks(ticks = np.arange(0, 12, 1), labels = list1, fontsize = 15)
plt.colorbar(fraction = 0.046, pad = 0.04)

plt.tight_layout()
plt.show()
```

## Analyse en composantes principales:

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
# réduction des variables
X=scale(spotify[list1])
```

```python
#définition de la commande
pcaR = PCA()
spotifyR = pd.DataFrame(X, columns = list1) #sur les données centrées réduites
#composantes principales
C = pcaR.fit(spotifyR).transform(spotifyR)
```

```python
plt.figure(figsize = (10,5))
x=np.arange(pcaR.explained_variance_ratio_.size)
plt.bar(x, pcaR.explained_variance_ratio_)
plt.axis((-1, 20, 0, 0.5))
plt.xlabel('Nombre de composoantes principales')
plt.ylabel('Variance expliquée')
plt.show()
#on garde 80 % de la variance expliquées donc on conserve 6 variables
```

```python
## distribution des composantes principales
plt.figure(figsize = (10, 10))
plt.boxplot(C[:,:])
plt.title("Boxplot des composantes principales")
plt.show()
```

```python
## Repésentation des individus
plt.figure(figsize=(10,10))
for i, j, classe in zip(-C[:,0], -C[:,1], spotify["pop.class"]):
    if classe=='A':
        color='red'
    elif classe=='B':
        color='blue'
    elif classe=='C':
        color='green'
    else:
        color='black'
    plt.plot(i, j, ".",color=color)
plt.axis((-10,10,-10,10))  
plt.show()
```

## Models

```python
spotify.head()
```

```python
# transformer les variables qualitatives en 0 et 1 (as factor R)
spotifyDum=pd.get_dummies(spotify[["key","mode"]])

spotifyQuant=spotify[["valence","year","acousticness","danceability","duration","energy","instrumentalness","liveness"
                      ,"loudness","speechiness","tempo"]]
dfC=pd.concat([spotifyDum,spotifyQuant],axis=1)
dfC.head()
```

```python
# variable à expliquer binaire
Popc=spotify["pop.class"]
# variable à expliquer réelle
Pop=spotify["popularity"]
```

```python
Pop.hist()
plt.show()
```

```python
from sklearn.model_selection import train_test_split  
X_train,X_test,Popc_train,Popc_test=train_test_split(dfC,Popc,test_size=200,random_state=11)
X_train,X_test,Pop_train,Pop_test=train_test_split(dfC,Pop,test_size=200,random_state=11)
```

```python
#Normalisation des données 
#indispensable pour beaucoup d'autres méthodes non linéaires (SVM, réseaux de neurones, modèles avec pénalisation).
from sklearn.preprocessing import StandardScaler  
# L'algorithme ds réseaux de neurones nécessite éventuellement une normalisation 
# des variables explicatives avec les commandes ci-dessous
scaler = StandardScaler()  
scaler.fit(X_train)  
Xr_train = scaler.transform(X_train)  
# Meme transformation sur le test
Xr_test = scaler.transform(X_test)
```

### Regression logistique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
#AIC
```

```python
# Optimisation du paramètre de pénalisation Lasso
# grille de valeurs
param=[{"C":[1,1.2,1.5,1.7,2,3,4]}]
logit = GridSearchCV(LogisticRegression(penalty="l1",solver="saga"), param,cv=5,n_jobs=-1) #"saga" préféré à "liblinear" for large datasets
logitOpt=logit.fit(Xr_train, Popc_train)  # GridSearchCV est lui même un estimateur
# paramètre optimal
logitOpt.best_params_["C"]
print("Meilleur score = %f, Meilleur paramètre = %s" % (1.-logitOpt.best_score_,logitOpt.best_params_))
```

```python
# erreur sur l'échantillon test
print("erreur:" +str(1-logitOpt.score(Xr_test, Popc_test)))
```

```python
# Prévision
Popc_chap= logitOpt.predict(Xr_test)
# matrice de confusion
table=pd.crosstab(Popc_chap,Popc_test)
print(table)
```
